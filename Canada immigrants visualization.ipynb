```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# Set the inline backend for Jupyter to display graphs within the notebook
%matplotlib inline

# Load the dataset
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
df_can = pd.read_csv(url, index_col='Country')

# Add a column to sum up total immigrants by country if it does not already exist
if 'Total' not in df_can.columns:
    df_can['Total'] = df_can.sum(axis=1)

years = list(map(str, range(1980, 2014)))
df_can.sort_values(['Total'], ascending=False, inplace=True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()

# Plotting
df_top5.plot(kind='area', stacked=False, figsize=(20, 10))  # Changed to 'stacked=False' for an unstacked area plot
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

# Use ggplot style for more aesthetically pleasing visuals
mpl.style.use('ggplot')

# Check Matplotlib version
print('Matplotlib version: ', mpl.__version__)  # >= 2.0.0

# Load dataset
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
df_can = pd.read_csv(url, index_col='Country')

# Show the first few entries of the dataset
print(df_can.head())

# Print the dimensions of the dataframe
print('Data dimensions:', df_can.shape)

# Set the country name as index - useful for quickly looking up countries using .loc method
df_can.index.name = None

# Let's create a list of years from 1980 - 2013, this will come in handy when plotting the data
years = list(map(str, range(1980, 2014)))

# Sorting by Total to get the top 5 countries
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()[years].transpose()
df_top5.index = df_top5.index.map(int)

# Generate Area Plot for the top 5 countries
df_top5.plot(kind='area', alpha=0.35, stacked=False, figsize=(20, 10))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()

# Generate Histogram
df_can['2013'].plot(kind='hist', figsize=(8, 5))
plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()

# Analyzing Iceland's immigration to Canada
df_iceland = df_can.loc['Iceland', years]
df_iceland.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')
plt.show()

# Annotating the bar graph with the financial crisis
df_iceland.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Icelandic Immigrants to Canada from 1980 to 2013')
plt.annotate('',                      # s: str. will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )
plt.show()

# Horizontal Bar Plot showing the total number of immigrants to Canada from the top 15 countries
df_top15 = df_can['Total'].head(15)
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Countries Contributing to Immigration to Canada from 1980 to 2013')
for index, value in enumerate(df_top15):
    label = format(int(value), ',')  # format int with commas
    plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Use ggplot style for aesthetics
plt.style.use('ggplot')

# Enable inline plotting for Jupyter Notebooks
%matplotlib inline

# Load the dataset
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
df_can = pd.read_csv(url, index_col='Country')

# Print the dimensions of the dataframe
print('Data dimensions:', df_can.shape)

# Set Country as the index for easier data manipulation
df_can.index.name = None

# Define a new dataframe for Denmark, Norway, and Sweden
df_dsn = df_can.loc[['Denmark', 'Norway', 'Sweden'], :]

# Calculate total values and proportions for Waffle Chart
total_values = sum(df_dsn['Total'])
category_proportions = df_dsn['Total'] / total_values

# Define Waffle Chart size
width = 40  # width of chart
height = 10  # height of chart
total_num_tiles = width * height  # total number of tiles

# Compute the number of tiles for each category
tiles_per_category = (category_proportions * total_num_tiles).round().astype(int)

# Initialize the Waffle Chart as an empty matrix
waffle_chart = np.zeros((height, width))

# Create a Waffle Chart
category_index = 0
tile_index = 0
for col in range(width):
    for row in range(height):
        tile_index += 1
        if tile_index > sum(tiles_per_category[:category_index + 1]):
            category_index += 1
        waffle_chart[row, col] = category_index

# Display the Waffle Chart
fig = plt.figure()
colormap = plt.cm.coolwarm
plt.matshow(waffle_chart, cmap=colormap)
plt.colorbar()

# Prettify with a legend
categories = df_dsn.index.values
values_cumsum = np.cumsum(df_dsn['Total'])
total_values = values_cumsum[-1]
legend_handles = [plt.matplotlib.patches.Patch(color=colormap(val/total_values), label=f"{cat} ({df_dsn['Total'][cat]})") for cat, val in zip(categories, values_cumsum)]
plt.legend(handles=legend_handles, loc='lower center', ncol=len(categories), bbox_to_anchor=(0.5, -0.1))
plt.show()

# Prepare data for Regression Plot
years = list(map(str, range(1980, 2014)))
df_total = pd.DataFrame(df_can[years].sum(axis=0))
df_total.index = map(int, df_total.index)
df_total.reset_index(inplace=True)
df_total.columns = ['year', 'total']

# Create Regression Plot with Seaborn
plt.figure(figsize=(10, 6))
sns.regplot(x='year', y='total', data=df_total, color='green')
plt.xlabel('Year')
plt.ylabel('Total Immigration')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.show()
```
